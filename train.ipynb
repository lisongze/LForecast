```python
import keras
import tensorflow as tf
import numpy as np

def pre_process_data(l_data):
    red_ball_data = np.zeros((len(l_data), 33))
    blue_ball_data = np.zeros((len(l_data), 16))
    
    for idx, v in enumerate(l_data):
        red_ball_data[idx][v[0]] = 1
        red_ball_data[idx][v[1]] = 1
        red_ball_data[idx][v[2]] = 1
        red_ball_data[idx][v[3]] = 1
        red_ball_data[idx][v[4]] = 1
        red_ball_data[idx][v[5]] = 1
        blue_ball_data[idx][v[6]] = 1
    
    return red_ball_data, blue_ball_data


print 'ok'
```

```python
from keras import layers
from keras.layers import Input, Dense, Conv1D, Concatenate, ConvLSTM2D
from keras.models import Model
from keras.layers import LSTM, Dropout

history_len = 32
RED_BALL_COUNT = 33
BLUE_BALL_COUNT = 16
red_inputs = Input(shape=(history_len, RED_BALL_COUNT))
blue_inputs = Input(shape=(history_len, BLUE_BALL_COUNT))
x1=LSTM(64, input_dim=RED_BALL_COUNT, input_length=history_len, return_sequences=True, dropout=0.5)(red_inputs)
x1=LSTM(64,return_sequences=False, dropout=0.5)(x1)

x2=LSTM(64, input_dim=BLUE_BALL_COUNT, input_length=history_len, return_sequences=True, dropout=0.5)(blue_inputs)
x2=LSTM(64, return_sequences=False, dropout=0.5)(x2)
print x2

x=Concatenate([x1, x2])
x=Dense(128, activation='relu', input_dim=128)(x)
x=Dense(256, activation='relu')(x)
x=Dense(64, activation='relu')(x)
predictions = Dense(RED_BALL_COUNT+BLUE_BALL_COUNT, 'softmax')

model = Model(inputs=[red_inputs, blue_inputs], outputs=predictions)
model.compile(optimizer='rmsprop', loss='mse')
```
