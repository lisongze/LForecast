```python
import xlrd
import pandas as pd
```

```python
in_excel = xlrd.open_workbook('/home/leemo/Documents/history.xlsx')
table = in_excel.sheets()[1]
nrows = table.nrows
ncols = table.ncols
print nrows, ncols

date_id = []
r1 = []
r2 = []
r3 = []
r4 = []
r5 = []
r6 = []
blue = []

for i in xrange(2, nrows):
    rowValues = table.row_values(i)
    date_id.append(int(rowValues[0]))
    r1.append(int(rowValues[1]))
    r2.append(int(rowValues[2]))
    r3.append(int(rowValues[3]))
    r4.append(int(rowValues[4]))
    r5.append(int(rowValues[5]))
    r6.append(int(rowValues[6]))
    blue.append(int(rowValues[7]))
#     print int(rowValues[0]), int(rowValues[1]), rowValues[2], rowValues[3], rowValues[4], rowValues[5], rowValues[6], rowValues[7]

date_id.reverse()
r1.reverse()
r2.reverse()
r3.reverse()
r4.reverse()
r5.reverse()
r6.reverse()
blue.reverse()

his_data = {'date_id':date_id, 'r1':r1, 'r2':r2, 'r3':r3, 
           'r4':r4, 'r5':r5, 'r6':r6, 'blue':blue}
pd_his = pd.DataFrame(his_data, columns=['date_id', 'r1', 'r2', 'r3', 'r4', 'r5', 'r6', 'blue'], index=range(nrows-2))
pd_his.to_csv('./history.csv')
print pd_his
```
