```python
import pandas as pd
import numpy as np
import scipy
from scipy.special import comb, perm
from enum import Enum

''' game state enumerate'''
class LGameState(Enum):
    GAME_ING = 1
    GAME_SUSPEND = 2
    GAME_OVER = 3

class LGameEngine:
    def __init__(self, init_money=10000000,
                 game_mode='normal',
                 red_ball_count=6,
                 blue_ball_count=1):
        self.game_mode = game_mode
        self.current_index = 0
        self.init_money = init_money
        self.game_red_ball_count = red_ball_count
        self.game_blue_ball_count = blue_ball_count
        self.current_money = self.init_money
        self.start_data_index = 0
        self.end_data_index = 0
    
    def load_l_csv(self, csv_file):
        self.l_data = pd.read_csv(csv_file)
        self.l_data_len = len(self.l_data.index)
    
    def get_l_data(self, start_index, end_index):
        try:
            data = self.l_data.loc[start_index:end_index,['r1', 'r2', 'r3', 'r4', 'r5', 'r6', 'blue']]
            data = data.as_matrix()
            return data
        except LookupError, e:
            print 'Has not load lottery data from csv', e
    
    def get_next_l_data(self, next_index):
        try:
            data = self.l_data.loc[next_index:next_index,['r1', 'r2', 'r3', 'r4', 'r5', 'r6', 'blue']]
            data = data.as_matrix()
            return data
        except IndexError, e:
            print 'Out of index', e
    
    def cal_bet_money(self, red_ball_count, blue_ball_count):
        if red_ball_count < self.game_red_ball_count:
            raise ValueError('Wrong red ball count !!!')
        if blue_ball_count < self.game_blue_ball_count:
            raise ValueError('Wrong blue ball count !!!')
        single_shot_money = 2
        cs = comb(red_ball_count, self.geme_red_ball_count) * blue_ball_count
        money = single_shot_money * cs
        return money
    
    def get_current_index(self):
        return self.current_index
    
    def set_current_index(self, index):
        self.current_index = index
    
    def set_game_data_range(self, start_data_index, end_data_index):
        self.start_data_index = start_data_index
        self.end_data_index = end_data_index
        
    def get_game_state(self):
        if self.current_money <= 0:
            return LGameState.GAME_OVER
        elif self.current_index >= self.end_data_index-1:
            return LGameState.GAME_SUSPEND
        else:
            return LGameState.GAME_ING;
    
    def bet(self, red_ball_pro, blue_ball_pro):    
        r_map = {}
        b_map = {}
        for idx, v in enumerate(red_ball_pro):
            r_map[idx] = v
        for idx, v in enumerate(blue_ball_pro):
            b_map[idx] = v
        r_map = sorted(r_map.items(), key=lambda d:d[1], reverse=True)
        b_map = sorted(b_map.items(), key=lambda d:d[1], reverse=True)     
        red_ball_order = [v[0] for v in r_map]
        blue_ball_order = [v[0] for v in b_map]
        
        print '-------red_ball_order:'
        print r_map
        print red_ball_order
        print '-------blue_ball_order:'
        print b_map
        print blue_ball_order
        
        bingo_num = self.get_next_l_data(self.current_index+1)[0]
        bingo_red_ball = bingo_num[:6]
        bingo_blue_ball = bingo_num[6]
        r_tmp = []
        b_tmp = []
        for idx, v in enumerate(bingo_red_ball):
            r_tmp.append(red_ball_order.index(v))
        b_tmp.append(blue_ball_order.index(bingo_blue_ball))
        
        bet_red_ball_count = max(r_tmp)+1
        bet_blue_ball_count = max(b_tmp)+1
        
        bet_money = self.cal_bet_money(bet_red_ball_count, bet_blue_ball_count)
        print 'bingo_num: ', bingo_num
        print 'r_tmp: ', r_tmp
        print 'b_tmp: ', b_tmp
        print 'bet_red_ball_count: ', bet_red_ball_count
        print 'bet_blue_ball_count: ', bet_blue_ball_count
```
